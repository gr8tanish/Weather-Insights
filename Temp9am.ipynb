```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
raw_df = pd.read_csv("weatherAUS.csv")
raw_df
```

```python
columns_drop = ['Rainfall', 'Evaporation', 'Sunshine', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'Cloud9am',	'Cloud3pm', 'RainToday', 'RainTomorrow']
raw_df = raw_df.drop(columns=columns_drop)
```

```python
raw_df['Temp9am'] = raw_df['Temp9am'].ffill()
```

```python
raw_df
```

```python
column_d = ['Temp3pm']
raw_df = raw_df.drop(columns=column_d)
```

```python
raw_df
```

```python
print(raw_df['Temp9am'].isna().sum())
```

```python
year = pd.to_datetime(raw_df.Date).dt.year

train_df = raw_df[year < 2015]
val_df = raw_df[year == 2015]
test_df = raw_df[year > 2015]
```

```python
input_cols = list(train_df.columns)[1:-1]
target_col = 'Temp9am'
```

```python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_col].copy()
```

```python
val_inputs = val_df[input_cols].copy()
val_targets = val_df[target_col].copy()
```

```python
test_inputs = test_df[input_cols].copy()
test_targets = test_df[target_col].copy()
```

```python
train_inputs
```

```python
numeric_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
categorical_cols = train_inputs.select_dtypes('object').columns.tolist()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean').fit(raw_df[numeric_cols])
```

```python
train_inputs[numeric_cols] = imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = imputer.transform(test_inputs[numeric_cols])
```

```python
test_inputs[numeric_cols].isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(raw_df[numeric_cols])
```

```python
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore').fit(raw_df[categorical_cols])
```

```python
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
encoded_cols
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

```python
test_inputs
```

```python
X_train = train_inputs[numeric_cols + encoded_cols]
X_val = val_inputs[numeric_cols + encoded_cols]
X_test = test_inputs[numeric_cols + encoded_cols]
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
```

```python
%%time
rf.fit(X_train, train_targets)
```

```python
rf.score(X_train, train_targets)
```

```python
rf.score(X_val, val_targets)
```

```python
rf.score(X_test, test_targets)
```

```python
model = RandomForestRegressor(
    n_estimators=300,
    max_depth=20,
    max_features='sqrt',
    random_state=3,
    n_jobs=-1
)
```

```python
model.fit(X_train, train_targets)
model.score(X_train, train_targets), model.score(X_val, val_targets)
```

```python
input_data = pd.DataFrame({
    'Location': ['Albury'],
    'MinTemp': [13.4],
    'MaxTemp': [22.9],
    'WindGustSpeed': [44],
    'WindSpeed9am': [20],
    'WindSpeed3pm': [24],
    'Humidity9am': [71],
    'Humidity3pm': [22],
    'Pressure9am': [1007.7],
    'Pressure3pm': [1007.1],
})
```

```python
input_data[numeric_cols] = imputer.transform(input_data[numeric_cols])

input_data[numeric_cols] = scaler.transform(input_data[numeric_cols])

encoded_input_data = encoder.transform(input_data[categorical_cols])

encoded_input_df = pd.DataFrame(encoded_input_data, columns=encoded_cols)

final_input_data = input_data[numeric_cols] + encoded_input_df

final_input_data = final_input_data.reindex(columns=X_train.columns)
```

```python
final_input_data
```

```python
prediction = model.predict(final_input_data)
```
